```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["font.family"] = "Malgun Gothic"  # 윈도우 한글 폰트
mpl.rcParams["axes.unicode_minus"] = False

# 예시 데이터 생성
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

# 가중치(계수)와 절편 확인
print(f"계수: {model.coef_}")
print(f"절편: {model.intercept_}")

# 예측 결과 시각화
plt.scatter(X, y, color="blue")
plt.plot(X, model.predict(X), color="red")
plt.xlabel("X")
plt.ylabel("y")
plt.title("선형 회귀 결과")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# 아이리스 데이터셋 로드
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# 결정 트리 모델 학습
model = DecisionTreeClassifier()
model.fit(X, y)

# 특성 중요도 확인
importances = model.feature_importances_
for name, importance in zip(feature_names, importances):
    print(f"{name}: {importance:.4f}")

# 특성 중요도 시각화
plt.barh(feature_names, importances)
plt.xlabel("특성 중요도")
plt.title("결정 트리 특성 중요도")
plt.show()
```

✅ 딥러닝에서 손실 함수, 활성화 함수, 옵티마이저가 역전파에 미치는 영향

1. 손실 함수 (Loss Function)
- 역할: 역전파의 시작점
- 설명:
  * 모델의 예측과 정답 사이의 차이를 계산
  * 이 차이(오차)를 기반으로 gradient(기울기)를 계산함
  * 손실 함수 없이는 역전파도 없음
- 영향:
  * 역전파에서 전파될 오차의 크기와 방향 결정

2. 활성화 함수 (Activation Function)
- 역할: gradient 전달의 필터
- 설명:
  * 각 뉴런의 출력값을 비선형적으로 변형
  * 역전파 시에는 이 함수의 미분값이 gradient에 곱해짐
- 영향:
  * gradient가 얼마나 잘 전달될지를 결정
  * 예: Sigmoid는 gradient를 작게 만들어 학습 느리게 함
  * 예: ReLU는 양수일 때 gradient를 잘 전달함

3. 옵티마이저 (Optimizer)
- 역할: 가중치를 실제로 업데이트
- 설명:
  * 역전파로 계산된 gradient를 바탕으로 가중치(weight)를 수정함
  * learning rate 등 전략을 통해 업데이트 방식 결정
- 영향:
  * 학습 속도와 안정성에 직접적 영향
  * 예: Adam은 빠르고 안정적, SGD는 단순하지만 느릴 수 있음

✅ 전체 흐름 요약

[순전파]
입력 → 가중치 계산 → 활성화 함수 → 예측값 → 손실 계산

[역전파]
손실 함수 → 오차 계산  
→ 활성화 함수의 미분값과 곱해서 gradient 계산  
→ 옵티마이저가 gradient를 사용해 가중치 업데이트

✅ 한 줄 요약
- 손실 함수: 오차의 출발점
- 활성화 함수: gradient 전달 조절자
- 옵티마이저: gradient를 실제로 적용하는 실행자
